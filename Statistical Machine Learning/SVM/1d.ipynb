```python
import numpy as np
from sklearn.svm import SVC
```

```python
svm = SVC(gamma='auto', kernel='linear')
X = np.array([[0,0],[1,2],[-1,2]])
y = np.array([-1,1,1])
```

```python
s = svm.fit(X,y)
```

```python
s.coef_
```

```python
s.intercept_
```

```python
s.dual_coef_
```

For the dual coef. there is minus in front as they are multiplied with y. Other than that, values are almost the same up to numerical imprecisions (e.g. w_0 is 0 in analytical but 0.000384)
